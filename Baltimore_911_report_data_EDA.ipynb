### Calculating the min, max, mean and median of the numeric data columns

```python
numeric_columns = ['Post', 'Longitude', 'Latitude', 'Total Incidents']

for column in numeric_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

statistics = df[numeric_columns].agg(['min', 'max', 'mean', 'median', 'std']).transpose()

print(statistics)
```

## Missingness in the rows

```python
total_rows = len(df)

missingness = df.isnull().sum() / total_rows * 100

print(missingness)
```
